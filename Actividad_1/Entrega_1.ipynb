# Second part of the analysis
Now I am going to do a coorrelation of the irrigation and the use of fertilizer with the type of crop.

```python
df_2 = pd.read_csv(path+ '\\' +file_2,sep=",",encoding='utf-8')
df_2

if not df_2.empty and 'Cultivo' in df_2.columns:
    # Asegúrate de que no haya datos nulos en la columna 'Cultivo'
    if df_2['Cultivo'].notnull().any():
        plt.figure(figsize=(10, 5))
        df_2['Cultivo'].value_counts().plot(kind='bar')
        plt.title('Distribución de tipos de cultivo')
        plt.xlabel('Cultivo')
        plt.ylabel('Frecuencia')
        plt.grid(axis='y')
        plt.show()
    else:
        print("La columna 'Cultivo' solo contiene valores nulos.")
else:
    print("El DataFrame está vacío o la columna 'Cultivo' no existe.")
```

```python
# Relationship between irrigation and type of crop
plt.figure(figsize=(10, 5))
cross_tab_riego_cultivo = pd.crosstab(df_2['Riego'], df_2['Cultivo'])
cross_tab_riego_cultivo.plot(kind='bar', stacked=True)
plt.title('Relación entre riego y tipo de cultivo')
plt.xlabel('Riego')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.legend(title='Cultivo')
plt.show()

# Relationship between fertilizers applied and type of crop
plt.figure(figsize=(10, 5))
cross_tab_fertilizante_cultivo = pd.crosstab(df_2['Fertilizantes aplicados'], df_2['Cultivo'])
cross_tab_fertilizante_cultivo.plot(kind='bar', stacked=True)
plt.title('Relación entre fertilizantes aplicados y tipo de cultivo')
plt.xlabel('Fertilizantes aplicados')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.legend(title='Cultivo')
plt.show()
```

```python
# Selecting relevant columns for mapping
relevant_columns = ['Riego', 'Fertilizantes aplicados', 'Cultivo']
correlation_data = df_2[relevant_columns]

# Convert categorical variables to numeric if necessary
correlation_data = pd.get_dummies(correlation_data, drop_first=True)

# Correlation matrix
correlation_matrix = correlation_data.corr()

# Correlation Matrix Display
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.title('Matriz de correlación')
plt.xticks(range(len(correlation_matrix)), correlation_matrix.columns, rotation=90)
plt.yticks(range(len(correlation_matrix)), correlation_matrix.columns)
plt.grid(False)
plt.show()
```

# Second part of the analysis

I am going to fin the relation with the gender and the master class and which class each gender tend to inscribe.

```python
# Gender distribution
plt.figure(figsize=(5, 5))
df_3['Genero'].value_counts().plot(kind='bar')
plt.title('Distribución de Género')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.show()

# Distribution of the Master class
plt.figure(figsize=(10, 5))
df_3['Master class'].value_counts().plot(kind='bar')
plt.title('Distribución de Master class')
plt.xlabel('Master class')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.show()
```

```python
# Cross table to show the relationship between Gender and Master class
cross_tab_genero_class = pd.crosstab(df_3['Genero'], df_3['Master class'])

# Stacked bar chart to show distribution
plt.figure(figsize=(10, 5))
cross_tab_genero_class.plot(kind='bar', stacked=True)
plt.title('Relación entre Género y Master class')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.grid(axis='y')
plt.legend(title='Master class')
plt.show()
```

```python
# Percentage of participation by gender in each Master class
cross_tab_percent = cross_tab_genero_class.div(cross_tab_genero_class.sum(axis=1), axis=0)

# Stacked Bar Chart with Percentages
plt.figure(figsize=(10, 5))
cross_tab_percent.plot(kind='bar', stacked=True)
plt.title('Porcentaje de participación por Género en cada Master class')
plt.xlabel('Género')
plt.ylabel('Porcentaje')
plt.grid(axis='y')
plt.legend(title='Master class')
plt.show()
```
